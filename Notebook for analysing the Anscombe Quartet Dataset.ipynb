## Background information on the Anscombe quartet dataset

The Anscombe’s Quarter dataset was constructed by Francis John Anscombe in 1973. *[1]*

*Francis Anscombe -* *[2]*
<img src="https://upload.wikimedia.org/wikipedia/en/d/d5/Francis_Anscombe.jpeg" width="200" height="200">

He intended to highlight the usefulness of graphs. He highlighted that graphs can have multiple purposes in order to assist in the perception of features of the data being analysed while also showing the alternate data that is contained in the dataset. Throughout the original article in "Graphs in Statistical Analysis". American Statistician. 27 (1): 17–21, which can be viewed [here](http://www.sjsu.edu/faculty/gerstman/StatPrimer/anscombe1973.pdf) he aimed to highlight that most kinds of statistical analysis on datasets rested on assumptions of the behaviour of the data in question which could in turn be false providing misleading calculations. *[3]*

Anscombe’s original article states that the dataset used was made up of four fictitious datasets of eleven x,y pairs which can be viewed below *[3]* - 

*Original Anscombe Quartet Dataset -* *[3]*



## Analysis of the Anscombe's quartet dataset

```python
# Importing packages required for analysis. 

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
```

The cell below shows that the dataset has been defined as df, as a short hand notation for further workings, and loaded in to the terminal.
The anscombe.csv file used in this analysis can also be viewed as part of this repository for review.

```python
df = pd.read_csv('anscombe.csv', usecols=['x1', 'x2', 'x3', 'x4', 'y1', 'y2', 'y3','y4'])
```

The print, head and tail functions print the first and last 5 columns of the dataset which is useful to ensure that the dataset has loaded correctly. As can be seen from printing df (the Anscombe dataset) the columns have a similar distance in range between the x and y valued columns in the four datasets. This can also be seen in the below statistics calculated that the mean, standard deviation, minimum value, maximum value and quartile values that the values are similar in each of the columns. *[3]*

### Summary statistic calculation using the ".describe" function in python to calculate the following - 

1) Count of the number of elements in each column

2) Mean

3) Standard deviation

4) Minimum Value

5) Interquartile range

6) Maximum Vaue

```python
# Print the characteristics for each column contained in the Anscombe's data set.

print(df.describe())
#print(df.count()) - not required as included in the .describe function but can be run separately in data analysis if a count is required
```

The above statistics are indicative of the similarities within the x valued columns, as well as the y valued columns. 

As can be seen from the above output table, the mean values and standard deviations are the same for each of the x valued columns within the dataset. This is also true for  the y valued columns. Such statistical calculations and outputs have the potential to cause an analyst to automatically assume or become biased to the fact that the grouped columns, x and y columns, are the same and therefore lead to incorrect assumptions during analysis.

### Calculation of the mean of the x and y values in the dataset

```python
# Code lines below combine all x value containing columns from the dataset into an array and calculates the mean which is correct as per the original analysis by Anscombe as the value computes to 9.

x2 = df[['x1', 'x2', 'x3', 'x4']].values
Meanx2 = np.mean(x2)
print('The mean of the X values in the dataset is', Meanx2)
```

```python
# Calculating the mean value of the y values

y2 = df[['y1', 'y2', 'y3', 'y4']].values
Meany2 = float (np.mean(y2))
print('The mean of the Y values in the dataset is', Meany2)

# Rounding of decimal places of the answer - to be investigated further to see if this can be printed to this value in previous print statement
# Calculation is correct as per original study.

print('This value can be rounded to', round(7.501, 2))
```

The above mean calculations are indicative of those mentioned is Anscombe's original analysis of the dataset. This indicates the the above code can accurately repeat the mean calculations completed by Anscombe. *[3]*

Such statistics as the mean and standard deviations are used in statistical analysis to provide an overall summary of the data contained in a dataset being investigated. However the following visualations will indicate that such statistics should not be wholy relied upon to give a full overview of the data collected during analysis, as suggested by Anscombe in his original article.

### Correlation and Covariance calculation -

```python
# print(df.corr()) - Prints correlation of all columns against each other

print( df[['x1','y1']].corr())
print( df[['x2','y2']].corr())
print( df[['x3','y3']].corr())
print( df[['x4','y4']].corr())
```

```python
#print(df.cov()) - Prints covariance of all columns against each other

print( df[['x1','y1']].cov())
print( df[['x2','y2']].cov())
print( df[['x3','y3']].cov())
print( df[['x4','y4']].cov())
```

The correlation coefficient value would generally indicate a mutual relationship or connection between the datasets if similar. While the covariance value indicates the property of a function of retaining its form when the variables are linearly transformed. As can be seen from the above calculations when each of the datasets are compared the result is leads to a determination of a similarity between the datasets. This relationship will be discussed further later.

## Visualising the dataset - 

### Overall view of the dataset - 

This plot provides an overall view of the information contained in the dataset plotted against the alternate columns in the dataset using various plots.

```python
# Pairplot - gives a general overview using various plots to give a generic picture of the data contained in the Anscombe's dataset.

sns.pairplot(df)
```

### Scatter plots of the x values plotted against their corresponding y values

The code lines in the cell below will produce a scatter plot of the x column against it's corresponding y value, such as x1 vs y1 etc. This code provides additional information to a generic scatter plot as it also includes the regression (or best fit line) for each of the plots produced. 

```python
#  LInear model regression line applied.
# Produces scatter plot for each x,y pairing contained within the dataset against one another. 
# Produces a better visualisation of the data over a generic scatter plot as it includes the regression line for analysis. [4]

sns.lmplot(x='x1', y='y1', data =df)
sns.lmplot(x='x2', y='y2', data =df)
sns.lmplot(x='x3', y='y3', data =df)
sns.lmplot(x='x4', y='y4', data =df)
```

A second series of scatter plots will follow which show better fit regression lines for each of the datasets as it is clear from the above that a linear regression model only appears to be applicable for the first dataset. This relationship will also be discussed further later.

```python
# Better fit model regression lines without the use of linear regression for each plot. [4]

sns.lmplot(x="x1", y="y1", data=df, ci=None, scatter_kws={"s": 80})
sns.lmplot(x="x2", y="y2", data=df, order=2, ci=None, scatter_kws={"s": 80})
sns.lmplot(x="x3", y="y3", data=df, ci=None, scatter_kws={"s": 80})
sns.lmplot(x="x4", y="y4", data=df, ci=None, scatter_kws={"s": 80})
```

### Bar plots produced when plotting information from all columns and individual plots of x values against their corresponding y values - 

The below plots provide additional visualisation, on top of the scatter plots below, of the differences between the 4 paired columns contained within the Anscombe dataset. 

```python
# Box plot of all columns combined in one diagram for comparison of the variation purposes.

df.plot.bar()
```

```python
# Individual plots to give a clearer picture than the above combined bar plot.
# Code below produces box plots of each x value plotted against its corresponding y value.

df.plot.bar(x='x1', y='y1')
df.plot.bar(x='x2', y='y2')
df.plot.bar(x='x3', y='y3')
df.plot.bar(x='x4', y='y4')
```

### Jointplot of each paired x and their corresponding y value -

The jointplot again gives additional visualisation of the scatter plots produced for each of the paired columns as well as the hisotgram distribution.

```python
# Produces histograms for each paired set in the dataset, x values v their corresponding y values.
# Also includes the scatter plot without regression line.

sns.jointplot(x="x1", y="y1", data=df)
sns.jointplot(x="x2", y="y2", data=df)
sns.jointplot(x="x3", y="y3", data=df)
sns.jointplot(x="x4", y="y4", data=df)
```

## Summary of analysis of the dataset

From the above analysis methods through both statistical calculation and visualisation it can be determined that although the data contained in Anscombe's quartet dataset appears quite similar from a statistical perspective, it is visualised in a different light that portrays different information to the statistical properties.This would be the main reasons for interest in Anscombe's dataset.

As indicated by the calculations above each of the datasets contained in Anscombe's Quartet Dataset have both similar mean and standard deviation values.  Each of the datasets has a mean value of approximately 9.0 and 7.5 for the X and Y valued columns respectively. While the standard deviations were calculated to be approximately 3.3 and 2.0 respectively. This would generally lead to the believe that each of the datasets involved would be similar.

This can also be seen by calculating the correlation coefficient and covariance seen above. Once each dataset was calculated it has produced values of approximately 0.82 for the correlation covarient and 5.5 for the covariance for each dataset. As each of the datasets have the same approximate values, similar to the mean and standard deviations, it would be indicative of a positive relationship between the datasets. 


As Anscombe states in his original article, the values used were ficitious. However, once analysed it is clear that this fictious dataset perfectly aligns with his argument that statistical calculations should not be solely used to define characteristics of a dataset being used in analysis. *[3]* 


Once producing plots of the datasets in Anscombe's quartet, based on the above statistics one would normally assume that the plots should also produce similar results due to an unconcious bias which can be based on such calculations. However, this is clearly not the case when reviewing the produced scatter plots above. The visualistions portrayed on the scatter plots and regression line show a very different aspect of the data. This visualisation was used by Anscombe in his study to prove his opinion that the statistical calculations would lead to incorrect assumptions when the data was visualised.


- Within the X1 Y1, first, scatter plot produced appears to have a linear shaped distribution of the datapoints. This can be seen in the case of both scatter plots produced as the best fit line in both cases remains as a linear relationship and doesn't alter that much compared to the re-plotting of the second dataset.


- The second scatter plot, X2 Y2, has a less normal distribution and rather appears as a curve once the datapoints have been plotted in such a manner. This required the second set of scatter plots to be produced which allows for the better visualisation of the non-linear trend within this dataset compared to the first set.


- The third scatter plot, X3 Y3, appears to have the greatest normal distribution of all 4 sets due to the linear relationship portrayed in this plot. However, the outlier data point must also be taken in to consideration as it impacts the calculations of the correlation coefficient, discussed earlier. Similarly to the re-plotting of the first dataset, the best fit line in both cases remains as a linear relationship and doesn't alter that much compared to the re-plotting of the second dataset.


- Finally, the fourth scatter plot, X4 Y4, has also been impacted by the outlier data point which can be visualised on both scatter plots produced. Excluding this data point, the remaining data points in this dataset appear to have no relationship between the variables within the plot due to the shape of the information once graphed in both manners.


From this the scatter plots provide a very different image to that which would have been expected if only the statistical aspects of the datasets were taken in to consideration. The plots provide a different portrayl of the data points contained in each of the datset and the fact that all four datasets provide very different information regarding the relationship of the variables once graphed. Not all of the datasets have a similar linear relationship as would be expected following the statistical analysis. The variance in the regression line plots also points to a very different outcome to that which would have been expected from the statistical calculations. Although the 4 datasets appear to have similar characteristics, the X4 Y4 scatter plot mentioned previously and shown above is clear that the regression line differs by a greater amount in this plot compared to plotting the data points from the previous 3 datasets. As can be seen for the various scatter plots and appropriate regression lines it is indicative of the different relationships within the datasets in Anscombe's quartet dataset which would not be assumed from the relationship evident from the calculated descriptive statistics.


This information is also backed up by the bar plots produced of the datasets, also produced earlier in the notebook. From reviewing the plots generated, the variance between the individual datasets is clearly evident and doesn't appear corroborate the similarities in the datasets which we were led to believe was present from the descriptive statistics calculated earlier in the notebook.


The above characteristics both enhance interest in Anscombe's quartet dataset for studies while providing a cautionary warning to be aware of a bias which can be introduced in to testing and analysis based solely on statistics.
